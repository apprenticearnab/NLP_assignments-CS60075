```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

x = raw.split("\n")
```

(Note: Please remove those which contain only articles, prepositions, determiners. For Example: “of the”, “in a”, etc).

```python
#Write code

from nltk.util import ngrams
unigrams=[]
bigrams=[]
trigrams=[]
fourgrams=[]

for content in (x):
    unigrams.extend(a for a in word_tokenize(content))
    bigrams.extend(list(ngrams(word_tokenize(content),2)))
    trigrams.extend(ngrams(word_tokenize(content),3))
    fourgrams.extend(ngrams(word_tokenize(content),4))
print(unigrams)
print(bigrams)
print(trigrams)
print(fourgrams)
    ##similar for trigrams and fourgrams
```

```python
#stopwords = code for downloading stop words through nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words1 = list(set(stopwords.words('english')))
stop_words2 = ['"','.',',','!','?',':',';']      #declaring array for deleting following punctuation marks
stop_words = stop_words1 + stop_words2

#print top 10 unigrams, bigrams after removing stopwords
uni_processed = [p for p in unigrams if p not in stop_words]
fdist1 = nltk.FreqDist(p for p in uni_processed)
mostCommon= fdist1.most_common(10)
print(mostCommon)

bi_processed = [tup for tup in bigrams if not False in [False for wrd in tup if wrd in stop_words]]
fdist2 = nltk.FreqDist(bi_processed)
mostCommon1= fdist2.most_common(10)
print(mostCommon1)

tri_processed = [tup for tup in trigrams if not False in [False for wrd in tup if wrd in stop_words]]
fdist3 = nltk.FreqDist(tri_processed)
mostCommon2= fdist3.most_common(10)
print(mostCommon2)

four_processed = [tup for tup in fourgrams if not False in [False for wrd in tup if wrd in stop_words]]
fdist4 = nltk.FreqDist(four_processed)
mostCommon3= fdist4.most_common(len(fdist4))
print(mostCommon3)

#print top 10 bigrams, trigrams, fourgrams after removing stopwords

```
